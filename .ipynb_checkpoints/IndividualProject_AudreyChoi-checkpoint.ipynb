## Implement Helper Functions

```python
def remove_punctuation(text):
    
    #replace new line(\n) with empty space
    text = text.replace("\n"," ")
    
    #remove all puncuation
    for char in punctuation:
        text = text.replace(char,"")
    
    #remove more puncuation that is not filtered by "puncuation"
    text = text.replace("’", " ")
    text = text.replace("“", " ")
    text = text.replace("”", " ")
    text = text.replace("–", " ")
    return text
```

```python
def remove_stopword(text):
    
    #get the list of stopwords in english
    stop = stopwords.words("english")
    
    #new empty list to append words that are not stopwords
    no_stops = []
    for word in text:
        if word not in stopwords.words('english'):
            no_stops.append(word)
            
    return no_stops
```

```python
def find_most_frequent_words(word_list,num):
    word_frequencies = Counter(word_list)
    most_frequent = word_frequencies.most_common(num)
    most_frequent_words = [word for word, _ in most_frequent]
    return most_frequent_words
```

```python
# Love letters generated by ChatGPT
print(open("Text/LoveLetter/Chat_GPT_Love_Letter1.txt",encoding="utf-8").read())
```

### Create a dictionary to map the name of text and text file path

```python
loveletter_text = {
"loveletter_ChatGPT1" : "Text/LoveLetter/Chat_GPT_Love_Letter1.txt",
"loveletter_ChatGPT2" : "Text/LoveLetter/Chat_GPT_Love_Letter2.txt",
"loveletter_ChatGPT3" : "Text/LoveLetter/Chat_GPT_Love_Letter3.txt",
"loveletter_1" : "Text/LoveLetter/Loveletter1.txt",
"loveletter_2" : "Text/LoveLetter/Loveletter2.txt",
"loveletter_3" : "Text/LoveLetter/Loveletter3.txt",
"loveletter_4" : "Text/LoveLetter/Loveletter4.txt",
"loveletter_5" : "Text/LoveLetter/Loveletter5.txt",
"loveletter_6" : "Text/LoveLetter/Loveletter6.txt",
"loveletter_7" : "Text/LoveLetter/Loveletter7.txt",
"loveletter_8" : "Text/LoveLetter/Loveletter8.txt",
"loveletter_9" : "Text/LoveLetter/Loveletter9.txt",
}
```

### Import file

```python
for key, value in loveletter_text.items():
    
    # Open all text files
    locals()[key] = open(value, encoding="utf-8").read()
    
    # Lowercasing
    locals()[key] = locals()[key].lower()
    
    #remove_punctuation_and_newlines
    locals()[key] = remove_punctuation(locals()[key])
    
    #Tokenization
    locals()[key]= locals()[key].split()
    
    #remove_stopword
    locals()[key] =remove_stopword(locals()[key])
```

### Create WordClouds

```python
love_ChatGPT_text = [loveletter_ChatGPT1, loveletter_ChatGPT2, loveletter_ChatGPT3]
love_ChatGPT_join = loveletter_ChatGPT1 + loveletter_ChatGPT2 + loveletter_ChatGPT3
loveletter_text = [
    loveletter_1,
    loveletter_2,
    loveletter_3,
    loveletter_4,
    loveletter_5,
    loveletter_6,
    loveletter_7,
    loveletter_8,
    loveletter_9
]

love_letter_join =loveletter_1+loveletter_2+loveletter_3+loveletter_4+loveletter_5+loveletter_6+loveletter_7+loveletter_8+loveletter_9
```

```python
# Create a WordCloud object
# Join the words back to create wordcloud
print ("Words from Love Letter generated by ChatGPT")
lovechatGPTcloud = WordCloud(width=800, height=200, background_color="white", colormap='Dark2').generate(' '.join(love_ChatGPT_join))

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(lovechatGPTcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Create a WordCloud object
# Join the words back to create wordcloud
print ("Words from real love letters")
lovelettercloud = WordCloud(width=800, height=200, background_color="white", colormap='Dark2').generate(' '.join(love_letter_join))

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(lovelettercloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Calculate sentiment_score of letters

```python
list_of_overall_sentiment_score_of_ChatGPT = []
for txt in love_ChatGPT_text:
    txt = ' '.join(txt)
    scores = sid.polarity_scores(txt)
    overall_sentiment_score = scores['compound']*100
    list_of_overall_sentiment_score_of_ChatGPT.append(overall_sentiment_score)
    
sentiment_score_of_chatGPT = list_of_overall_sentiment_score_of_ChatGPT
print("sentiment_score_of_chatGPT: ",sentiment_score_of_chatGPT)
average_sentiment_score_of_chatGPT = sum(sentiment_score_of_chatGPT)/len(sentiment_score_of_chatGPT)
```

```python
list_of_overall_sentiment_score_of_loveletter = []
for txt in loveletter_text:
    txt = ' '.join(txt)
    scores = sid.polarity_scores(txt)
    overall_sentiment_score = scores['compound']*100
    list_of_overall_sentiment_score_of_loveletter.append(overall_sentiment_score)
    
sentiment_score_of_loveletter = list_of_overall_sentiment_score_of_loveletter
print("sentiment_score_of_loveletter: ",sentiment_score_of_loveletter)
average_sentiment_score_of_loveletter = sum(sentiment_score_of_loveletter)/len(sentiment_score_of_loveletter)
```

```python
# Generate x-axis values for ChatGPT and love letters
x_chatGPT = range(1, len(sentiment_score_of_chatGPT) + 1)
x_loveletters = range(1, len(sentiment_score_of_loveletter) + 1)

# Create a scatter plot
plt.scatter(x_loveletters, sentiment_score_of_loveletter, label='Love Letters')
plt.scatter(x_chatGPT, sentiment_score_of_chatGPT, label='ChatGPT')
plt.xlabel('Sample Index')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Score Comparison')

plt.legend()
plt.show()
```

```python
labels = ['ChatGPT', 'Love Letters']
scores = [average_sentiment_score_of_chatGPT, average_sentiment_score_of_loveletter]

plt.bar(labels, scores)
plt.ylabel('Sentiment Score')
plt.title('Average Sentiment Score Comparison')

plt.show()
```

## Analysis

All versions of love letters generated by ChatGPT receive high sentiment scores comparable to those of real love letters. The average sentiment score of ChatGPT-generated letters is slightly higher than that of the real love letters. These findings highlight the remarkable ability of ChatGPT to generate emotionally engaging text. However, further research and evaluation are necessary to explore the full potential and limitations of AI-generated love letters in capturing the intricacies and nuances of human emotions.

# Conclusion

In summary of the four evaluations of the writing ability of AI systems, it is observed that AI can translate text quite accurately, but faces challenges when it comes to liberal translation compared to literal translation. This is because AI can translate words by direct mapping, but struggles to interpret the actual meaning of the given words within a context.
Furthermore, AI is capable of generating informative news articles; however, the quality heavily relies on the available information. If insufficient information is provided, the generated text may lack depth and informativeness.
Surprisingly, AI can also produce emotional letters using sentimental words. It is important to note that this does not imply that AI possesses emotions. Instead, it demonstrates the AI's ability to mimic the nuances of human emotion.
Overall, these evaluations shed light on both the strengths and limitations of AI in various writing tasks, highlighting the need for further advancements to enhance the accuracy, contextual understanding, and depth of AI-generated content.


In Digital Humanities, AI's writing abilities offer valuable insights. Collaborating with human experts, AI can unlock new possibilities for creative expression and interdisciplinary exploration. By combining AI's strengths with human expertise, we advance understanding and application in the digital age. The synergy between AI and the humanities holds immense potential for innovation.
